```python
# Importowanie wymaganych bibliotek
import dask.dataframe as dd
import pickle
import pandas as pd
from joblib import load
```

```python
# Ładowanie danych z pliku Parquet za pomocą Dask
df = dd.read_parquet('output_directory/*.parquet')

# Dodajemy kolumnę 'hour' z godziny z 'tpep_pickup_datetime'
df['hour'] = df['tpep_pickup_datetime'].dt.hour

# Losowo wybieram 10% rekordów
df_sample = df.sample(frac=0.00001, random_state=42)
df_sample = df_sample.compute()

# Cechy i target
X_sample = df_sample[['trip_distance', 'fare_amount', 'passenger_count', 'hour', 'payment_type', 'RatecodeID', 'trip_duration']]
y_sample = df_sample['tip_amount']
```

```python
# Ładowanie modelu z pliku
with open('models/lightgbm_regressor_model.pkl', 'rb') as f:
    lightgbm_regressor_model = pickle.load(f)

y_pred_lightgbm_regressor_model = lightgbm_regressor_model.predict(X_sample)
```

```python
# Ładowanie modelu z pliku
with open('models/linear_regression_model.pkl', 'rb') as f:
    linear_regression_model = pickle.load(f)

y_pred_linear_regression_model = linear_regression_model.predict(X_sample)
```

```python
# Ładowanie modelu z pliku
with open('models/xgboost_regressor_model.pkl', 'rb') as f:
    xgboost_regressor_model = pickle.load(f)

y_pred_xgboost_regressor_model = xgboost_regressor_model.predict(X_sample)
```

```python
# Ładowanie modelu z pliku
with open('models/random_forest_regressor_model.pkl', 'rb') as f:
    random_forest_regressor_model = pickle.load(f)

y_pred_random_forest_regressor_model = random_forest_regressor_model.predict(X_sample)
```

```python
# DataFrame z porównaniem
comparison_df = pd.DataFrame({
    'Rzeczywista wartość': y_sample,
    'LightGBM (Przewidywana)': y_pred_lightgbm_regressor_model,
    'Linear Regression (Przewidywana)': y_pred_linear_regression_model,
    'XGBoost (Przewidywana)': y_pred_xgboost_regressor_model,
    'Random Forest 1% (Przewidywana)': y_pred_random_forest_regressor_model
    # 'Random Forest 100% (Przewidywana)': y_pred_random_forest_regressor_model_joblib
})

# Wyświetlenie rekordów
print(comparison_df.to_string(index=False, line_width=200))
```
